```python
import random as rd
import time

inicio=time.time()


lista=[]

for i in range(0,100000):
    n=rd.randint(0000,1000000)
    lista.append(n)


def implis(lista:list):
    for i in range(0,len(lista)):
        print(lista[i])


def ord(lista:list)->list:
    for i in range (0,len(lista)):
        for z in range(0,len(lista)-1):
            if lista[z]>lista[z+1]:
                aux=lista[z]
                lista[z]=lista[z+1]
                lista[z+1]=aux
    return lista


implis(lista)
ot=ord(lista)
implis(ot)
fin=time.time()
print(" ")
print(fin-inicio)
```
